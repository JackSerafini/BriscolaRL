```python
import re
import matplotlib.pyplot as plt
import numpy as np

# File containing the training logs
LOG_FILE = "outputs/output_200000_dqn.txt"

# Lists to store parsed values
episodes = []
mean_rewards = []
win_rates = []
recent_win_rates = []
epsilons = []

# Regex pattern matching your log format
pattern = re.compile(
    r"Episode\s+(\d+)/\d+\s+\|\s+Mean reward \(last \d+\):\s+([+-]?\d+\.\d+)\s+\|\s+"
    r"Win rate:\s+(\d+\.\d+)%\s+\|\s+Win rate \(last \d+\):\s+(\d+\.\d+)%\s+\|\s+"
    r"Epsilon:\s+(\d+\.\d+)"
)

# Read and parse file
with open(LOG_FILE, "r") as file:
    for line in file:
        match = pattern.search(line)

        if match:
            episodes.append(int(match.group(1)))
            mean_rewards.append(float(match.group(2)))
            win_rates.append(float(match.group(3)))
            recent_win_rates.append(float(match.group(4)))
            epsilons.append(float(match.group(5)))

print(f"Parsed {len(episodes)} episodes.")
```

### Plotting

```python
fig, ax = plt.subplots(figsize=(12, 6))

ax.plot(episodes, win_rates, label="Win Rate (%)", linewidth=2)
ax.plot(episodes, recent_win_rates, label="Recent Win Rate (%)")
ax.set_xlabel("Episodes")
ax.set_ylabel("Win Rate")
ax.grid(True)

ax.set(ylim=(0, 100), yticks=np.arange(0, 101, 10))
ax.legend()

# ax.plot(x2, y2 + 2.5, 'x', markeredgewidth=2)
# ax.plot(x, y, linewidth=2.0)
# ax.plot(x2, y2 - 2.5, 'o-', linewidth=2)

# ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
#        ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```

```python
fig, ax1 = plt.subplots(figsize=(12, 6))

# Mean rewards
ax1.plot(episodes, mean_rewards, label="Mean Reward", linewidth=2)
ax1.set_xlabel("Episodes")
ax1.set_ylabel("Mean Reward")
ax1.grid(True)

# Win rates on secondary axis
ax2 = ax1.twinx()

ax2.plot(
    episodes,
    win_rates,
    label="Win Rate (%)",
    linestyle="--"
)

ax2.plot(
    episodes,
    recent_win_rates,
    label="Recent Win Rate (%)",
    linestyle=":"
)

ax2.set_ylabel("Win Rate (%)")

# Combine legends
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()

ax1.legend(lines1 + lines2, labels1 + labels2, loc="best")

plt.title("Training Progress")
plt.tight_layout()
plt.show()

# -------------------------
# Optional epsilon plot
# -------------------------

# plt.figure(figsize=(12, 4))
# plt.plot(episodes, epsilons)
# plt.xlabel("Episodes")
# plt.ylabel("Epsilon")
# plt.title("Epsilon Decay")
# plt.grid(True)
# plt.tight_layout()
# plt.show()
```
