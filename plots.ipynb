```python
import re
import matplotlib.pyplot as plt
import numpy as np
```

```python
DQN_LOG = "outputs/output_200000_dqn.txt"
PPO_LOG1 = "outputs/output_500000_ppo.txt"
PPO_LOG2 = "outputs/output_1000000_ppo.txt"
```

```python
episodes_dqn = []
mean_rewards_dqn = []
win_rates_dqn = []
recent_win_rates_dqn = []
epsilons_dqn = []

pattern = re.compile(
    r"Episode\s+(\d+)/\d+\s+\|\s+"
    r"Mean reward \(last \d+\):\s+([+-]?\d+\.\d+)\s+\|\s+"
    r"Win rate:\s+(\d+\.\d+)%\s+\|\s+"
    r"Win rate \(last \d+\):\s+(\d+\.\d+)%\s+\|\s+"
    r"Epsilon:\s+(\d+\.\d+)"
)

with open(DQN_LOG, "r") as file:
    for line in file:
        match = pattern.search(line)

        if match:
            episodes_dqn.append(int(match.group(1)))
            mean_rewards_dqn.append(float(match.group(2)))
            win_rates_dqn.append(float(match.group(3)))
            recent_win_rates_dqn.append(float(match.group(4)))
            epsilons_dqn.append(float(match.group(5)))

print(f"Parsed {len(episodes_dqn)} episodes.")
```

```python
episodes_ppo1 = []
mean_rewards_ppo1 = []
win_rates_ppo1 = []
recent_win_rates_ppo1 = []
value = []

pattern = re.compile(
    r"Episode\s+(\d+)/\d+\s+\|\s+"
    r"Mean reward \(last \d+\):\s+([+-]?\d+\.\d+)\s+\|\s+"
    r"Win rate:\s+(\d+\.\d+)%\s+\|\s+"
    r"Win rate \(last \d+\):\s+(\d+\.\d+)%\s+\|\s+"
    r"Policy loss:\s+([+-]?\d+\.\d+)\s+\|\s+"
    r"Value loss:\s+([+-]?\d+\.\d+)\s+\|\s+"
    r"Entropy:\s+([+-]?\d+\.\d+)"
)

with open(PPO_LOG1, "r") as file:
    for line in file:
        match = pattern.search(line)

        if match:
            episodes_ppo1.append(int(match.group(1)))
            mean_rewards_ppo1.append(float(match.group(2)))
            win_rates_ppo1.append(float(match.group(3)))
            recent_win_rates_ppo1.append(float(match.group(4)))
            value.append(float(match.group(6)))

print(f"Parsed {len(episodes_ppo1)} episodes.")
```

```python
episodes_ppo2 = []
mean_rewards_ppo2 = []
win_rates_ppo2 = []
recent_win_rates_ppo2 = []
# epsilons = []

pattern = re.compile(
    r"Episode\s+(\d+)/\d+\s+\|\s+"
    r"Mean reward \(last \d+\):\s+([+-]?\d+\.\d+)\s+\|\s+"
    r"Win rate:\s+(\d+\.\d+)%\s+\|\s+"
    r"Win rate \(last \d+\):\s+(\d+\.\d+)%\s+\|\s+"
    r"Policy loss:\s+([+-]?\d+\.\d+)\s+\|\s+"
    r"Value loss:\s+([+-]?\d+\.\d+)\s+\|\s+"
    r"Entropy:\s+([+-]?\d+\.\d+)"
)

with open(PPO_LOG2, "r") as file:
    for line in file:
        match = pattern.search(line)

        if match:
            episodes_ppo2.append(int(match.group(1)))
            mean_rewards_ppo2.append(float(match.group(2)))
            win_rates_ppo2.append(float(match.group(3)))
            recent_win_rates_ppo2.append(float(match.group(4)))
            # epsilons.append(float(match.group(5)))

print(f"Parsed {len(episodes_ppo2)} episodes.")
```

### Plotting

```python
fig, ax = plt.subplots(figsize=(12, 6))

ax.plot(episodes_dqn, win_rates_dqn, label="DQN Win Rate (%)", linewidth=2, color='g')
ax.plot(episodes_dqn, recent_win_rates_dqn, label="DQN Recent Win Rate (%)", color='g', alpha=0.5)

ax.set_xlabel("Episodes")
ax.set_ylabel("Win Rate")
ax.grid(True)

ax.set(ylim=(40, 100), yticks=np.arange(40, 101, 10),
       xlim=(0, 200000), xticks=np.arange(0, 200001, 20000))
ax.legend()

plt.title("Training Progress")
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12, 6))

ax.plot(episodes_ppo1, win_rates_ppo1, label="PPO Win Rate (%)", linewidth=2, color='b')
ax.plot(episodes_ppo1, recent_win_rates_ppo1, label="PPO Recent Win Rate (%)", color='b', alpha=0.5)

ax.set_xlabel("Episodes")
ax.set_ylabel("Win Rate")
ax.grid(True)

ax.set(ylim=(40, 100), yticks=np.arange(40, 101, 10),
       xlim=(0, 500000), xticks=np.arange(0, 500001, 25000))
ax.legend()

plt.title("Training Progress")
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12, 6))

ax.plot(episodes_ppo2, win_rates_ppo2, label="PPO Win Rate (%)", linewidth=2, color='r')
ax.plot(episodes_ppo2, recent_win_rates_ppo2, label="PPO Recent Win Rate (%)", color='r', alpha=0.5)

ax.set_xlabel("Episodes")
ax.set_ylabel("Win Rate")
ax.grid(True)

ax.set(ylim=(40, 100), yticks=np.arange(40, 101, 10),
       xlim=(0, 1000000), xticks=np.arange(0, 1000001, 50000))
ax.legend()

plt.title("Training Progress")
plt.tight_layout()
plt.show()
```
