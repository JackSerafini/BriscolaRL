```python
import random

from briscola import Briscola
```

```python
environment = Briscola()
environment.reset()
```

```python
observations, info = environment.reset()
```

```python
terminated = False
total_reward = 0
```

```python
while not terminated:
    action = random.randint(0, len(environment.player_hand) - 1)

    observations, reward, terminated, _, info = environment.step(action)
    total_reward += reward

    print(f"Action: {action}")
    print(f"Observation: {observations}")
    print(f"Reward: {reward}")
```

```python
print(f"Total reward: {total_reward}")
```
